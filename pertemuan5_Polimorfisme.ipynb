##POLIMORFISME##

Dari segi bahasa polimorfisme tediri dari dua bahasa latin, yaitu poly yang berarti banyak dan morph yang mepunyai arti bentuk. jadi polimorfisme berati banyak bentuk.

Didalam pemrograman objek, polimorfisme adalah konsep dimana terdapat beberapa class yang mempunyaimethod yang sama. Untuk penerapan dari method-method tersebut adalah diserahkan kepada tiap kelas yang yang memiliki method, akan tetapi untuk cara pemanggilan harus sama.

Cara pengaksesan Polimorfisme dibedakan menjadi dua jenis yaitu:
1. Polimorfisme dengan menggunakan fungsi
2. Polimorfisme degan menggunakn Method


```python
#Contoh Polimorfisme

class Shark:
    def swim(self):
        print("the shark is swimming")
    def swim_backwards(self):
        print ("the shark cannot swim backwards, but can sink backwards")
    def skeleton(self):
        print ("the shark's skeleton is made of cartilage")
    
class Clownfish:
    def swim(self):
        print ("the clownfish is swimming")
    def swim_backwards(self):
        print ("the clownfis can swim backwards")
    def skeleton(self):
        print ("the clownfish's skeloeton is made of bone.")

#cara 1(biasa)
print ('##mengakses dengan cara biasa##')
sammy=Shark()
sammy.skeleton()

casey=Clownfish()
casey.skeleton()
print ()


#cara 2(function)
print ('##mengakses dengan menggunakan fungsi##')
def sea_animal(ikan):
    ikan.swim()
    ikan.swim_backwards()
sammy=Shark()
casey=Clownfish()

sea_animal(sammy)
sea_animal(casey)
print ()

#cara 3(method)
print ('##mengakses dengan menggunakan method##')
sammy=Shark()
cassey=Clownfish()
for fish in (sammy,cassey):
    fish.swim()
    fish.swim_backwards()
    fish.skeleton()
```

Penjelasan contoh program:

Dalam contoh program diatas pertama kita membuat dua class dengan nama Shark dan Clownfish, dimana di masing-masing class tersebut mempunyai method dengan nama method yang sama antara keduanya yaitu methode swim, swim_backwards, dan skeleton tetapi masing-masing methode mempunyai nilai yang berbeda di setiap classnya.

Untuk pemanggilan method tiap kelas ada 3 cara:
1. menggunakan cara biasa
Dengan menggunakan cara ini kita menggunakan cara manual, yaitu dengan membuatkan objek setiap class yang ada dan kemudian memanggimethod yang ada di class tersebut.

2. mengunakan fungsi polimorfisme
Untuk memanggil method dengan cara ini kita membut fungsi yang bernama sea_animal yang mempunyai parameter bernama ikan, kemudian parameter ikan tersebut kita gunakan untuk memanggil method class dengan cara: parameter.namaMethod
Kemudian kita membuat objek untuk mendeklarasikan masing-masing class, disini kita menggunakan nama sammy untuk class Shark dan cassey untuk class Clownfish.
Dan untuk mengeksekusi fungsi polimorfisme tersebut kita cukup menuliskan nama fungsi dengan nama paramater objek yang telah kita buat sebelumnya di main program, dan dengan cara itu maka class yang ada di atas methodnya sudah akan berjalan dan menampilkan nilai sesuai apa yang di miliki masing-masing method.

3. menggunakan method polomorfisme
Dengan menggunakan method polimorfisme, langkah pertama adalah membut objek untuk mendeklarasikan nama masing-masing class.
Selanjutnya adalah mebuat looping secara iterasi menggunkaan parameter objek yang telah kita buat sebelumnya, hal ini dimaksukan agar nantinya berjala berurutan sesuai urutan parameter yang di masukkan.
selanjutnya yaitu di dalam looping tersebut kita panggil method yang ingin kita eksekusi, sehingga nanti akan mengeksekusi dan memunculkan nilai dari method tersebut secara berurutan.

```python
#latihan 1

class pegawai:
    def __init__(self,nama):
        self.nama=nama
    def Gaji(self):
        raise NotImplementedError("sub class smust implement abstract methode")

class manager(pegawai):
    def __init__(self,nama,gaji,tunjangan):
        self.gaji=gaji
        self.tunjangan=tunjangan
        pegawai.__init__(self,nama)
    def Gaji(self):
        print ('gaji Rp.',self.gaji)
    def gaji_plus(self):
        print ('gaji + tunjangan',self.gaji+ self.tunjangan)
class programmer(pegawai):
    def __init__(self,nama,gaji,bonus):
        self.gaji=gaji
        self.bonus=bonus
        pegawai.__init__(self,nama)
    def Gaji(self):
        print ('gaji Rp.',self.gaji)
    def gaji_plus(self):
        print ('gaji + bonus',self.gaji + self.bonus)

print ('##menggunakan sfunction##')
def kantor(kerja):
    print ('Nama: ',kerja.nama)
    kerja.Gaji()
    kerja.gaji_plus()

pekerja1=manager('jono',8000000,2000000)
pekerja2=programmer('jini',5000000,1500000)

kantor(pekerja1)
kantor(pekerja2)
print ()

print ('##menggunakan metode##')
for kerja in (pekerja1,pekerja2):
    print ('Nama:',kerja.nama)
    kerja.Gaji()
    kerja.gaji_plus()
```

Contoh dari program di atas adalah Polimorfisme class Abstrac. Dimana pada class pegawai di atas tidak mempunyai method untuk mengimplementasikan nilai dari class pegawai tersebut, akan tetapi hanya mendefinisikan struktur yang harus dimiliki oleh class yang lain sebagai bentuk polimorfisme.
Seperti contoh class di bawahnya yairu programmer dan kantor masing-masing harus memiliki method gaji sebagai struktur yang telah ada di class pegawai di atas.
Untuk cara kerja dan cara pemanggilan dari class yang berjenis abstrac ini tidak ada perbedaan denga yang lain, bisa menggunakan fungsi atau method tergantung mana yang dirasa lebih mudah.

```python
#latihan 2 (tanpa inharitance)
class manager:
    def __init__(self,nama,gaji,tunjangan):
        self.gaji=gaji
        self.tunjangan=tunjangan
        self.nama=nama
    def Gaji(self):
        print ('gaji Rp.',self.gaji)
    def gaji_plus(self):
        print ('gaji + tunjangan',self.gaji+ self.tunjangan)
class programmer:
    def __init__(self,nama,gaji,bonus):
        self.gaji=gaji
        self.bonus=bonus
        self.nama=nama
    def Gaji(self):
        print ('gaji Rp.',self.gaji)
    def gaji_plus(self):
        print ('gaji + bonus',self.gaji + self.bonus)

def kantor(kerja):
    print ('Nama: ',kerja.nama)
    kerja.Gaji()
    kerja.gaji_plus()

pekerja1=manager('jono',8000000,2000000)
pekerja2=programmer('wewe',5000000,1500000)

kantor(pekerja1)
kantor(pekerja2)
print ()

print ('##menggunakan metode##')
for kerja in (pekerja1,pekerja2):
    print ('Nama:',kerja.nama)
    kerja.Gaji()
    kerja.gaji_plus()
```

Yang dimaksud tanpa inharitance dalam program diatas adalah tidak adanya keterkaitan antara class yang satu dengan class yang lain dalam penggunaan atribut.
